```r
install.packages("tidyverse")
install.packages("docxtractr")
install.packages("ggfortify")
install.packages("gridExtra")

suppressMessages(library(tidyverse))
library(docxtractr)
library(ggfortify)
library(gridExtra)
```

```r
populations <- docx_extract_all_tbls(read_docx("https://ndownloader.figshare.com/files/10657603"))[[1]]
names(populations) <- populations[1,]
populations <- populations[-1,-c(3, 4)]
```

```r
populations <-
  populations %>%
  separate_rows(`Genetic distance`, sep = ', ')
names(populations)[2] <- "Sample"
```

```r
populations
```

```r
data_values <- docx_extract_all_tbls(read_docx("https://ndownloader.figshare.com/files/10657609"))[[1]]
```

```r
names(data_values)[1] <- "Sample"
```

```r
all_data <- inner_join(populations, data_values, by = "Sample")
```

```r
all_data
```

```r
pop_codes <- c(
  "Dongxiang" = "DX",
  "Fogang" = "FG",
  "Boluo" = "BL",
  "Zengcheng" = "ZC",
  "Huilai" = "HL",
  "Gaozhou" = "GZ",
  "Suixi" = "SX",
  "Qinghai" = "QH"
)
```

```r
p1 <- ggplot(all_data, aes(x = Population, y = as.numeric(`Plant height (cm)`))) +
  stat_boxplot(geom = "errorbar") +
  geom_boxplot() + scale_x_discrete(labels = pop_codes) +
  scale_y_continuous(limits = c(40, 220)) +
  theme_bw() +
  theme(axis.title.x = element_blank(), panel.grid = element_blank()) +
  labs(y = "plant height (cm)")

p2 <- ggplot(all_data, aes(x = Population, y = as.numeric(`Panicle length (cm)`))) +
  stat_boxplot(geom = "errorbar") +
  geom_boxplot() + scale_x_discrete(labels = pop_codes) +
  scale_y_continuous(limits = c(0, 350)) +
  theme_bw() +
  theme(axis.title.x = element_blank(), panel.grid = element_blank()) +
  labs(y = "panicle length (cm)")

p3 <- ggplot(all_data, aes(x = Population, y = as.numeric(`Awn length (mm)`))) +
  stat_boxplot(geom = "errorbar") +
  geom_boxplot() + scale_x_discrete(labels = pop_codes) +
  scale_y_continuous(limits = c(0, 140)) +
  theme_bw() +
  theme(axis.title.x = element_blank(), panel.grid = element_blank()) +
  labs(y = "awn length (mm)")

p4 <- ggplot(all_data, aes(x = Population, y = as.numeric(`Spikelet length (mm/ten spikelets)`))) +
  stat_boxplot(geom = "errorbar") +
  geom_boxplot() + scale_x_discrete(labels = pop_codes) +
  scale_y_continuous(limits = c(40, 160)) +
  theme_bw() +
  theme(axis.title.x = element_blank(), panel.grid = element_blank()) +
  labs(y = "spikelet length (mm)")

p5 <- ggplot(all_data, aes(x = Population, y = as.numeric(`Spikelet width (mm/ten spikelets)`))) +
  stat_boxplot(geom = "errorbar") +
  geom_boxplot() + scale_x_discrete(labels = pop_codes) +
  scale_y_continuous(limits = c(15, 40)) +
  theme_bw() +
  theme(axis.title.x = element_blank(), panel.grid = element_blank()) +
  labs(y = "spikelet width (mm)")

p6 <- ggplot(all_data, aes(x = Population, y = as.numeric(`Spikelet length-width ratio`))) +
  stat_boxplot(geom = "errorbar") +
  geom_boxplot() + scale_x_discrete(labels = pop_codes) +
  scale_y_continuous(limits = c(2, 6)) +
  theme_bw() +
  theme(axis.title.x = element_blank(), panel.grid = element_blank()) +
  labs(y = "spikelet length to width ratio")

p7 <- ggplot(all_data, aes(x = Population, y = as.numeric(`Number of spikelets per panicle`))) +
  stat_boxplot(geom = "errorbar") +
  geom_boxplot() + scale_x_discrete(labels = pop_codes) +
  scale_y_continuous(limits = c(0, 160)) +
  theme_bw() +
  theme(axis.title.x = element_blank(), panel.grid = element_blank()) +
  labs(y = "spikelet per panicle")

p8 <- ggplot(all_data, aes(x = Population, y = as.numeric(`Filled spikelets per panicle`))) +
  stat_boxplot(geom = "errorbar") +
  geom_boxplot() + scale_x_discrete(labels = pop_codes) +
  scale_y_continuous(limits = c(0, 100)) +
  theme_bw() +
  theme(axis.title.x = element_blank(), panel.grid = element_blank()) +
  labs(y = "filled spikelet per panicle")

p9 <- ggplot(all_data, aes(x = Population, y = as.numeric(`Secondary branches`))) +
  stat_boxplot(geom = "errorbar") +
  geom_boxplot() + scale_x_discrete(labels = pop_codes) +
  scale_y_continuous(limits = c(0, 18)) +
  theme_bw() +
  theme(axis.title.x = element_blank(), panel.grid = element_blank()) +
  labs(y = "secondary branches")

p10 <- ggplot(all_data, aes(x = Population, y = as.numeric(`Seed set (%)`))) +
  stat_boxplot(geom = "errorbar") +
  geom_boxplot() + scale_x_discrete(labels = pop_codes) +
  scale_y_continuous(limits = c(0, 120)) +
  theme_bw() +
  theme(axis.title.x = element_blank(), panel.grid = element_blank()) +
  labs(y = "seed set (%)")

ggsave("zac_ence_plot.pdf", grid.arrange(p1, p2, p3, p4, p5, p6, p7, p8, p9, p10, ncol = 2), height = 15, width = 7)
```

```r
filter(all_data, as.numeric(`Seed set (%)`) > 100)
```
